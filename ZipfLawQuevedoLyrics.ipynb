```python
# Dealing with system
import os
# Plotting
import matplotlib.pyplot as plt
# Handling txt files
import numpy as np
# Estimating frequency of elements in list
import collections
```

# Define util function

```python
def read_n_clean(fname):
  # Read raw data
  with open(fname) as file:
    lyrics = file.readlines()

  # Join all elements in one chunk and We do not want to be case sensitive
  lyrics = ' '.join(lyrics).lower()

  # Remove line beak \n
  lyrics = lyrics.replace('\n', '')

  # Remove spurious symbols (after visual inspection)
  lyrics = lyrics.replace('(', '').replace(')','').replace(',','')

  # Remove ''', DO NOT REMOVE ACCENTS (this is spanish and happy with accents)
  lyrics = lyrics.replace('\'', '')

  # Split words
  words_in_lyrics = lyrics.split()

  # print(words_in_lyrics)
  return(words_in_lyrics)
```

```python
def extract_frequency(words_in_lyrics):
  # Count frequency of elements
  frequency = collections.Counter(words_in_lyrics)
  # Sort by freq and split quants
  words = list()
  freq = np.array([])
  for w in sorted(frequency, key=frequency.get, reverse=True):
      words=np.append(words,w)
      freq=np.append(freq,frequency[w])

  return(words,freq)
```

# Checking Zipf law in just one song

```python
fname='dame_quevedo.txt'
words_in_lyrics = read_n_clean(fname=fname)
words,freq=extract_frequency(words_in_lyrics=words_in_lyrics)

# Plotting top-30 words
freqred = freq[:30]
wordsred = words[:30]
plt.figure(figsize=(28,7))
fig = plt.plot(np.arange(len(freqred)),freqred,lw=3.0,color='k')
fig = plt.ylabel('frequency',fontsize=25)
fig = plt.xlim([0,len(freqred)])
# plt.loglog()

# Compute max frequency to scale
freqmax = max(freqred)
print(f'~~ Maximum frequency = {freqmax}')
# Print the number of unique words
print(f'~~ Number of unique words = {len(words)}')
# plt.text(20,10000,f'Num. of words {len(words)}',fontsize=15)
plt.text(20,20,'$\mathcal{P}_n \sim 1/n^a$ with $n \in [1,158]$',fontsize=25)

# Following Wikipedias prescription try "a a bit higher than 1"
for a in np.linspace(0.5,1.5,10):
  # a = 1.5
  n = np.arange(len(freqred))
  prob = 1/n**a
  # Scaling Zipf law with max freq in data
  fig = plt.plot(n,prob*freqmax,ls='--',lw=2.0,label=f'a={np.round(a,2)}')

plt.title('DAME song by Quevedo',fontsize=30)
fig = plt.xticks(ticks=np.arange(len(freqred)),labels=wordsred,rotation=-45,fontsize=15)
fig = plt.yticks(fontsize=15)
plt.legend(fontsize=15,frameon=False)
plt.savefig('song.png',
            dpi=300,
            bbox_inches='tight')
```
